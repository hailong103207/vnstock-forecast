```python

# query_ohlcv_grouped: kết quả dạng dict[resolution][symbol] → DataFrame (index = Date)
data = query.query_ohlcv_grouped(
    symbols=["SHB", "VHM"],
    resolutions=["D", "W"],
    from_ts=1747008000,
    to_ts=1747267200,
)

# Duyệt và hiển thị từng (resolution, symbol)
for resolution, symbols_dict in data.items():
    for symbol, df in symbols_dict.items():
        print(f"\n{'='*50}")
        print(f"  Resolution: {resolution}  |  Symbol: {symbol}  |  {len(df)} rows")
        print(f"{'='*50}")
        display(df)
```
